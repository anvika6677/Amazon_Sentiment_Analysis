```python
def missing_values_analysis(df):
  na_columns_ = [col for col in df.columns if df[col].isnull().sum() >0]
  n_miss = df[na_columns_].isnull().sum().sort_values(ascending=True)
  ratio_=(df[na_columns_].isnull().sum() / df.shape[0]* 100).sort_values(ascending=True)
  missing_df=pd.concat([n_miss,np.round(ratio_, 2)],axis=1, keys=['Missing Values', 'Ratio'])
  return missing_df

def check_dataframe(df,head=5, tail=5):
  print("SHAPE".center(82,'~'))
  print("Rows: {}".format(df.shape[0]))
  print('columns: {}'.format(df.shape[1]))
  print("TYPES".center(82,'~'))
  print(df.dtypes)
  print("".center(82,'~'))
  print(missing_values_analysis(df))
  print("DUPLICATED VALUES".center(82,'~'))
  print(df.duplicated().sum())
  print("QUANTILES".center(82,'~'))
  numeric_df = df.select_dtypes(include=[float, int])
  print(numeric_df.quantile([0, 0.05, 0.50, 0.95, 0.99, 1]).T)


check_dataframe(df)

```

```python
def check_class(dataframe):
  nunique_df = pd.DataFrame({'Variable': dataframe.columns,
                            'Classes':[dataframe[i].nunique() \
                                       for i in dataframe.columns]})
  nunique_df = nunique_df.sort_values('Classes', ascending = False)
  nunique_df = nunique_df.reset_index(drop = True)
  return nunique_df

check_class(df)
```

```python



constraints = ['#834D22', '#EBE00C', '#1FEB0C', '#0C92EB', '#E80CD5']

def categorical_variable_summary(df, column_name):
    fig = make_subplots(rows=1, cols=2,
                        subplot_titles=('Countplot', 'Percentage'),
                        specs=[[{"type": "xy"}, {"type": "domain"}]])

    # Bar plot
    fig.add_trace(go.Bar(
        y=df[column_name].value_counts().values.tolist(),
        x=[str(i) for i in df[column_name].value_counts().index],
        text=df[column_name].value_counts().values.tolist(),
        textfont=dict(size=14),
        name=column_name,
        textposition='auto',
        showlegend=False,
        marker=dict(color=constraints, line=dict(color='#DBECEC', width=1))
    ), row=1, col=1)

    # Pie chart
    fig.add_trace(go.Pie(
        labels=df[column_name].value_counts().keys(),  # Fixed typo: 'lables' to 'labels'
        values=df[column_name].value_counts().values,
        textfont=dict(size=20),
        showlegend=False,
        name=column_name,
        marker=dict(colors=constraints)
    ), row=1, col=2)

    # Update layout
    fig.update_layout(title={
        'text': column_name,
        'y': 0.9,
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    }, template='plotly_white')

    # Display the figure
    fig.show()  # Use fig.show() instead of iplot()

# Example usage:
# categorical_variable_summary(df, 'column_name')
```

```python
import re
import pandas as pd

# Check if review_example is missing or not a string
if pd.isna(review_example) or not isinstance(review_example, str):
    cleaned_review = ""
else:
    # Step 1: Remove punctuations and numbers using regular expressions
    cleaned_review = re.sub(r'[^a-zA-Z\s]', '', review_example)

    # Step 2: Convert to lowercase
    cleaned_review = cleaned_review.lower()

    # Step 3: Split words and print each on a new line
    words = cleaned_review.split()

    # Step 4: Print each word on a new line
    for word in words:
        print(word)


```

```python
rt=lambda x: re.sub("[^a-zA-Z]",' ',str(x))
df["reviewText"]=df["reviewText"].map(rt)
df["reviewText"] = df["reviewText"].str.lower()
df.head()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

df[['polarity','subjectivity']] = df['reviewText'].apply(lambda Text:pd.Series(TextBlob(Text).sentiment))

# Use .items() to iterate over the Series index and values
for index, row in df['reviewText'].items():

  score = SentimentIntensityAnalyzer().polarity_scores(row)

  neg=score['neg']

  neu=score['neu']

  pos=score['pos']

  if neg>pos:

    df.loc[index,'sentiment']="Negative"

  elif pos>neg:

    df.loc[index,'sentiment']="Positive"

  else:

    df.loc[index,'sentiment']="Neutral"
```
